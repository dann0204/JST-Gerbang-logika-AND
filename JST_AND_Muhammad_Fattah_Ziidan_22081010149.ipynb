PROGRAM JARINGAN SARAF TIRUAN LOGIKA AND

```python
import random

# Fungsi Aktivasi (step function)
def fungsi_aktivasi(x):
    return 1 if x > 0 else 0  # Mendefinisikan fungsi aktivasi (step function) yang mengembalikan 1 jika input > 0, selain itu mengembalikan 0.

# Implementasi Jaringan Saraf Tiruan
class NN:
    def __init__(self):
        self.aktivasi = 0  # Inisialisasi variabel aktivasi.
        self.error = 0  # Inisialisasi variabel error.
        self.weights = [random.random() for _ in range(3)]  # Inisialisasi bobot dengan nilai acak untuk tiga bobot.
        self.miu = 0.1  # Inisialisasi nilai learning rate.
        self.no = 1  # Inisialisasi nomor iterasi.

    # Metode untuk menghitung output
    def hitung_output(self, x1, x2):
        self.aktivasi = self.weights[0] + x1 * self.weights[1] + x2 * self.weights[2]  # Menghitung nilai aktivasi dengan kombinasi linear dari input dan bobot.
        return fungsi_aktivasi(self.aktivasi)  # Mengembalikan output berdasarkan fungsi aktivasi.

    # Metode untuk melakukan pembelajaran
    def latih(self, x1, x2, target):
        output = self.hitung_output(x1, x2)  # Menghitung output dari input.
        self.error = target - output  # Menghitung error sebagai selisih antara target dan output.

        # Mencetak nilai-nilai selama proses pembelajaran
        print(f"{self.no:<4}{1:<3}{x1:<3}{x2:<3}{self.weights[0]:<8.3f}{self.weights[1]:<8.3f}"
              f"{self.weights[2]:<8.3f}{self.aktivasi:<10.3f}{output:<7}{target:<10}{self.error:<5}")

        # Mengupdate bobot jika error tidak sama dengan 0
        if self.error != 0:
            self.weights[0] += self.miu * self.error
            self.weights[1] += self.miu * self.error * x1
            self.weights[2] += self.miu * self.error * x2

        self.no += 1  # Meningkatkan nomor iterasi.

def main():
    total_sukses = 0  # Inisialisasi jumlah sukses untuk menghitung keberhasilan pembelajaran.
    max_epochs = 1000  # Inisialisasi jumlah maksimum epoch.
    AND = NN()  # Membuat instance dari kelas NN.

    # Persiapan
    inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]  # Mendefinisikan input untuk gerbang AND.
    target_output = [0, 0, 0, 1]  # Mendefinisikan output target untuk gerbang AND.

    # Melakukan Pembelajaran
    # Mencetak header untuk tabel hasil pembelajaran
    print(f"{'no':<4}{'1':<3}{'x1':<3}{'x2':<3}{'w1':<8}{'w2':<8}{'w3':<8}{'aktivasi':<10}"
          f"{'output':<7}{'target':<10}{'error':<5}")

    while total_sukses < 4 and AND.no // 4 < max_epochs:  # Melakukan pembelajaran sampai semua kasus berhasil atau mencapai maksimum epoch.
        print(f"\nEpoch : {AND.no // 4 + 1}")  # Mencetak nomor epoch saat ini.
        for i in range(4):  # Iterasi melalui semua kombinasi input.
            x1, x2 = inputs[i]
            target = target_output[i]

            AND.latih(x1, x2, target)  # Melatih jaringan dengan pasangan input dan target.

            # Memeriksa apakah pembelajaran sukses untuk kasus saat ini.
            if AND.error == 0 and x1 == 0 and x2 == 0:
                total_sukses = 1
            elif AND.error == 0:
                total_sukses += 1
            else:
                total_sukses = 0

    # Mencetak bobot akhir dan jumlah epoch yang diperlukan
    print(f"\nw1 = {AND.weights[0]}\nw2 = {AND.weights[1]}\nw3 = {AND.weights[2]}\nepoch = {AND.no // 4}")
    print("NN AND SELESAI!!!")

if __name__ == "__main__":
    main()  # Memanggil fungsi main untuk menjalankan program.
```

UNTUK MENGECEK APAKAH BOBOT YANG DIBERIKAN PADA PROGRAM SEBELUMNYA EMANG BENAR DAN DAPAT DIGUNAKAN UNTUK PEMBELAJARAN YANG LAIN.

```python
import random

# Fungsi Aktivasi (step function)
def fungsi_aktivasi(x):
    return 1 if x > 0 else 0

# Implementasi Jaringan Saraf Tiruan
class NN:
    def __init__(self, weights=None):
        self.aktivasi = 0
        self.error = 0
        self.weights = weights if weights else [random.random() for _ in range(3)]
        self.miu = 0.1
        self.no = 1

    # Metode untuk menghitung output
    def hitung_output(self, x1, x2):
        self.aktivasi = self.weights[0] + x1 * self.weights[1] + x2 * self.weights[2]
        return fungsi_aktivasi(self.aktivasi)

    # Metode untuk melakukan pembelajaran
    def latih(self, x1, x2, target):
        output = self.hitung_output(x1, x2)
        self.error = target - output
        print(f"{self.no:<4}{1:<3}{x1:<3}{x2:<3}{self.weights[0]:<8.3f}{self.weights[1]:<8.3f}"
              f"{self.weights[2]:<8.3f}{self.aktivasi:<10.3f}{output:<7}{target:<10}{self.error:<5}")

        # Mengupdate bobot
        if self.error != 0:
            self.weights[0] += self.miu * self.error
            self.weights[1] += self.miu * self.error * x1
            self.weights[2] += self.miu * self.error * x2

        self.no += 1

def main():
    # Input bobot secara manual atau menggunakan bobot yang dihasilkan dari program sebelumnya
    # Misalnya, bobot dari program sebelumnya adalah: [-0.4597209004101962, 0.21034567157686787, 0.36433873733868627]
    weights = [-0.4597209004101962, 0.21034567157686787, 0.36433873733868627]
    AND = NN(weights)

    # Persiapan data
    inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]
    target_output = [0, 0, 0, 1]

    # Menampilkan header tabel
    print(f"{'no':<4}{'1':<3}{'x1':<3}{'x2':<3}{'w1':<8}{'w2':<8}{'w3':<8}{'aktivasi':<10}"
          f"{'output':<7}{'target':<10}{'error':<5}")

    # Melakukan pembelajaran hingga mencapai konvergensi atau max_epochs
    total_sukses = 0
    max_epochs = 1000
    while total_sukses < 4 and AND.no // 4 < max_epochs:
        print(f"\nEpoch : {AND.no // 4 + 1}")
        for i in range(4):
            x1, x2 = inputs[i]
            target = target_output[i]
            AND.latih(x1, x2, target)

            if AND.error == 0:
                total_sukses += 1
            else:
                total_sukses = 0

    # Menampilkan bobot akhir dan jumlah epoch yang diperlukan
    print(f"\nw1 = {AND.weights[0]}\nw2 = {AND.weights[1]}\nw3 = {AND.weights[2]}\nepoch = {AND.no // 4}")
    print("NN AND SELESAI!!!")

    # Menguji dengan data baru menggunakan bobot yang telah dilatih
    print("\nPengujian dengan bobot yang dilatih:")
    for i in range(4):
        x1, x2 = inputs[i]
        output = AND.hitung_output(x1, x2)
        print(f"Input: ({x1}, {x2}), Output: {output}, Target: {target_output[i]}")

if __name__ == "__main__":
    main()
```
